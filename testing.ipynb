```python
import yfinance as yf

ticker = yf.Ticker("SBIN.NS")
news = ticker.news

titles = []
dates = []
for i in range(len(news)):
    titles.append(news[i]['content']['title'])
    dates.append(news[i]['content']['pubDate'])
```

```python
from combined_prediction_pipeline import CombinedPredictionPipeline
predictor = CombinedPredictionPipeline()
```

```python
HEADLINE = titles[2]
DATE = dates[2]
print(HEADLINE)
print(DATE)

predictor.predict(
    headline=HEADLINE,
    date=DATE,
    ticker="HDFCBANK"
)
```
